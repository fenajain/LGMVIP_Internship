**Methodology: Applying Sketch Filter to an Image**

**1. Importing Packages**

The code begins by importing the necessary packages for the image processing task. Two important packages are imported: `cv2` and `matplotlib.pyplot`. 

- `cv2` (OpenCV): A popular computer vision library used for image processing tasks, including reading, manipulating, and saving images.
- `matplotlib.pyplot`: A plotting library that provides a MATLAB-like interface for creating visualizations.

**2. Reading the Image**

The original image is read using the `cv2.imread()` function, which takes the image file path as an argument and returns a NumPy array representing the image.

**3. Applying the Sketch Filter**

The sketch filter is applied to the image in several steps:

- Convert to grayscale: The `cv2.cvtColor()` function is used to convert the image from the BGR color space to grayscale (`cv2.COLOR_BGR2GRAY`). This step simplifies the image by removing color information.
- Invert the grayscale image: The `cv2.bitwise_not()` function is used to invert the grayscale image, effectively reversing the pixel values.
- Apply Gaussian blur: The `cv2.GaussianBlur()` function applies a Gaussian blur to the inverted grayscale image. This blurring reduces noise and smoothens the image. The blur size is defined by the kernel size, which is set to (101, 101) in this case.
- Invert the blurred image: Similar to the earlier step, the `cv2.bitwise_not()` function is used to invert the blurred image.
- Create sketch filter: The `cv2.divide()` function is used to create the sketch filter. It divides the grayscale image by the inverted blurred image, applying a specific scaling factor (`scale=256.0`). This step enhances the edges and details in the image, producing a sketch-like effect.

**4. Converting the Filtered Image to Color**

To display the sketch-filtered image in color, the grayscale image is converted back to the BGR color space using `cv2.cvtColor()` with the flag `cv2.COLOR_GRAY2BGR`. This step allows the filtered image to be visually presented alongside the original image.

**5. Displaying the Images**

The code creates a figure and two axes using `plt.subplots(1, 2)`, which generates a single row with two columns for displaying the original and filtered images side by side. 

- Original image display: The original image is displayed on the left axis (`axs[0]`) using `axs[0].imshow()`, which takes the image array converted to RGB format.
- Filtered image display: The filtered image is displayed on the right axis (`axs[1]`) using `axs[1].imshow()` after converting it to RGB format.
- Axis configuration: The axes are configured to turn off the axis labels (`axs[0].axis('off')` and `axs[1].axis('off')`) and set titles for each image (`axs[0].set_title()` and `axs[1].set_title()`).
- Adjusting subplot spacing: `plt.tight_layout()` is used to adjust the spacing between the subplots, ensuring they are properly displayed.
- Showing the figure: Finally, `plt.show()` is called to display the figure with both images.

This methodology explains the steps involved in applying a sketch filter to an image using OpenCV and Matplotlib, along with descriptions of the packages and functions used.

### Importing the required libraries

```python
# Installing the required libraries
import cv2
from matplotlib import pyplot as plt
```

### Reading the image file

```python
# Read the original image
image = cv2.imread('dog.jpeg')
```

### Displaying the original image file

```python
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(rgb_image)
plt.title("Original Image")
plt.show()
```

### Converting the original image to the grayscale image

```python
# Apply the sketch filter
grey_filter = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
invert = cv2.bitwise_not(grey_filter)
```

```python
blur = cv2.GaussianBlur(invert, (101, 101), 0)
```

```python
inverted_blur = cv2.bitwise_not(blur)
```

```python
sketch_filter = cv2.divide(grey_filter, inverted_blur, scale=256.0)
```

```python
# Convert the sketch-filtered image to color
sketch_filter_color = cv2.cvtColor(sketch_filter, cv2.COLOR_GRAY2BGR)
```

### Saving the sketch image

```python
# Saving the sketch file in the png format
sketch = cv2.imwrite("output.png", sketch_filter_color)
```

### Plotting both the images

```python
# Create a figure and axes to display the images
fig, axs = plt.subplots(1, 2)

# Display the original image on the left
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].axis('off')
axs[0].set_title('Original Image')

# Display the sketch-filtered image on the right
axs[1].imshow(cv2.cvtColor(sketch_filter_color, cv2.COLOR_BGR2RGB))
axs[1].axis('off')
axs[1].set_title('Sketch Filtered Image')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the figure with both images
plt.show()
```
